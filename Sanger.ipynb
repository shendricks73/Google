```python
import folium

# Coordinates for Sanger, CA (approximate center)
latitude = 36.7097
longitude = -119.5576

# Create a map centered at Sanger, CA
m = folium.Map(location=[latitude, longitude], zoom_start=13)

# Add a marker for Sanger, CA
folium.Marker([latitude, longitude], popup='Sanger, CA').add_to(m)

# Display the map
m
```
